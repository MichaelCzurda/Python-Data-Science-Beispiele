### Allgemeines
KNN ist ein Supervised Machine Learning Algorithmus der verwendet werden kann um Regressionen und KLassifizierungen durchzuführen. Folgend werden die Begriffe oberflächlich erklärt um eine Grundlage für die in diesem Dokument betrachteten Konzepte zu schaffen.

#### Supervised Machine Learning
Für das sogenannte „überwachte Lernen“ werden bekannte Daten benötigt, die bereits die Logik beinhalten, die man gerne auf ein neues Datenset anwenden möchte. Ein Datenset ist in diesem Fall eine tabellenförmige Struktur, bestehend aus Spalten und Zeilen. Eine Zeile stellt eine Beobachtung und eine Spalte eine Variable dieser Einheit dar. Die einzelnen Untersuchungen bestehen aus unabhängigen Variablen("predictors") und einer abhängigen Variable("Label"). 

Der Algorithmus lernt die Logik innerhalb dieser Daten. Ein so trainierter Algorithmus kann die gelernte Logik anschließend auf "Predictors" neuer Daten anwenden, um den Wert des "Labels" zu ermitteln.

#### Klassifizierung
Bei einem Klassifizierungsproblem ist die zu erklärende Variable ein diskreter Wert. Die Klassen können dabei zB. die binäre Unterscheidung von "wahr"
und "falsch" sein, oder eine Unterteilung in verschiedene Klassen sein, wie etwa die Zuordnung von Menschen du ihrer ethnischen Gruppe.

#### Regression
Ein Regressionsproblem hat eine reelle Zahl als zu ermittelnden Wert. Dies könnte zB die Ermittlung des Gewichts eines Menschen sein, anhand der Informationen Geschlecht und Körpergröße.
